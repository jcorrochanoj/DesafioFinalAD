```python
import pandas as pd
import re
import numpy as np
```

```python
url = "datos_negocios_limpios.csv"
df = pd.read_csv(url)
df["categorias"]=df["categorias"].apply(lambda s: eval(s))
df=df.drop(list(df.columns)[0],axis=1)
```

```python
categorias=[]
for i in range(192609):
    categorias+=df.iloc[i]["categorias"]
print(categorias.count("Golf"))
for i in range(len(categorias)):
    categorias[i]=categorias[i].strip()
categorias = list(set(categorias))    
```

```python
listaDeCategorias=[]
for i in range(len(categorias)):
    listaDeCategorias.append([])
```

```python
for i in range(192609):
    au=df.iloc[i]["categorias"]
    aux=[]
    for j in range(len (au)):
        aux.append(au[j].strip())
    for k in range(len(categorias)):
        if (categorias[k] in aux):
            listaDeCategorias[k].append(1)
        else:
            listaDeCategorias[k].append(0)
```

```python
for i in listaDeCategorias:
    if(len(i)!=192609):
        print("problemo")
print(len(categorias))
print(len(listaDeCategorias))
```

```python
array=np.array(listaDeCategorias)
df2 = pd.DataFrame(array.transpose(),index=range(192609),columns=categorias)
```

```python
for i in categorias:
    df[i]=df2[i]
df=df.drop("categorias",axis=1)
del(df2)
df["Golf"].sum()
```

# Explicacion del codigo
Este codigo tiene como objetivo extraer conocimiento del atributo categorias, convirtiendo las diferentes categorias que existen en atributos binarios, para ello se ha creado una lista donde se analizaba cuantas categorias existen, despues de ello se ha creado un dataframe con tantas filas como  el original donde se ha ido marcando como 0 o 1 en las columnas en funcion de si en el dataframe original tenian esas caracteristicas o no, y finalmente se ha incluido este dataframe en el original como columnas extra, eliminando la columna de categorias original.
