### 2.1. ¿Qué negocios son tendencia en el último mes en Las Vegas?

El objetivo de este análisis es evitar usar la media de puntuación global de los negocios para ver realmente cuáles son los negocios en tendencia actualmente. Utilizaremos la media de puntuación de un periodo de tiempo corto y actual y el número de opiniones para encontrar los negocios en tendencia actualmente. 

Para ello, necesitaremos atender a los datos de un periodo corto de tiempo, en concreto, al último mes de opiniones. Para ello, obtendremos cuál es la ultima opinión de la base de datos para conocer cual es la última fecha del conjunto de opiniones, y a partir de ella, quedarnos con las opiniones realizadas en ese periodo de tiempo para los negocios de una zona, en este caso lo haremos sobre Las Vegas.

### Paso 1: Obtención de la última fecha recogida por opiniones

```python
import pandas as pd
opiniones = pd.read_csv("../Dataset/datos_opiniones_sintexto.csv")
opiniones = opiniones.rename(columns={'puntuacion':'puntuacion_opi'})
```

```python
opiniones.sort_values(['anyo', 'mes', 'dia'], ascending=False).head(10)
```

La fecha de las ultimas opiniones recogidas es del día 14 de Noviembre de 2018, por lo que nos quedaremos con las opiniones que estén entre las fechas 14 de Octubre y 14 de Noviembre de 2018.

### Paso 2: Obtención de las opiniones del último mes

```python
opiniones = opiniones[(opiniones['anyo'] == 2018) & (((opiniones['mes'] == 10) & (opiniones['dia'] >= 14)) | ((opiniones['mes'] == 11) & (opiniones['dia'] <= 14)))]
```

```python
opiniones
```

Solo en el último mes se han recogido una cantidad de más de 100000 opiniones. Ya que solo nos queremos centrar en una ciudad, lo haremos sobre la ciudad de Las Vegas. Para ello necesitamos cruzar nuestras opiniones con los negocios y obtener solo los de las vegas.

El proceso completo será:

1. Obtener los negocios de Las Vegas
2. Unir opiniones con negocios para tener todas las opiniones del ultimo mes de los negocios de Las Vegas.
3. Obtener la media de las puntuaciones de los negocios y el número de opiniones por negocio.
4. Unir estos datos con los negocios de nuevo para tener la ubicacion de cada negocio.

### Paso 3: Obtención de puntuacion media del ultimo mes y numero de opiniones

```python
negocios = pd.read_csv("../Dataset/datos_negocios.csv")
negocios = negocios.rename(columns={'puntuacion':'puntuacion_neg'})

# SOLO NOS QUEDAMOS CON LOS NEGOCIOS ABIERTOS

negocios = negocios[negocios['abierto'] == 1]
```

```python
df = pd.merge(opiniones, negocios, on='negocio_id', how='inner')
```

```python
lasVegas = df[df['estado'] == "NV"]
```

```python
lasVegas = lasVegas.rename(columns = {'puntuacion_opi':'punt_med_ult_mes'})
punt_med = lasVegas.groupby('negocio_id').mean()['punt_med_ult_mes'].to_frame()

count_opi = lasVegas.groupby('negocio_id').count()['opinion_id'].to_frame()
count_opi = count_opi.rename(columns = {'opinion_id':'count'})

lasVegas = pd.merge(punt_med, count_opi, on='negocio_id', how='inner')
```

```python
lasVegas
```

```python
lasVegas = pd.merge(lasVegas, negocios, on='negocio_id', how='inner')
lasVegas
```

### Paso 4: Graficar los datos en un mapa

Ahora que ya tenemos los datos de los negocios, es hora de ver los negocios en un mapa. Antes de verlos en un mapa, ordenaremos los negocios por numero de opiniones y puntuación para ver en una tabla cuál de ellos parece estar en tendencia en este último mes.

Decidimos ordenarlo en primer lugar por número de opiniones ya que lo consideramos un buen ponderador de la puntuación que tiene el negocio. No es lo mismo un negocio que tiene una puntuación de 5 estrellas con una opinion, que uno con 5 estrellas y 100 opiniones.

Decidimos quedarnos con los 20 mejores negocios para no saturar el mapa y ver los realmente importantes.

```python
lasVegas = lasVegas.sort_values(['count', 'punt_med_ult_mes'], ascending=False).head(20)
lasVegas.head(5)
```

Al ver los resultados, sin duda algo debemos estar haciendo bien ya que el restaurante que aparece en tendencia en las fechas en las que hemos elegido es el restaurante de Gordon Ramsay, un Chef muy conocido.

Con estos datos no solo podemos ver los que son tendencia en ese momento, sino que también podemos fijarnos en la puntuación media del negocio (puntuacion_neg) y compararla con la media del ultimo mes (punt_med_ult_mes). Mirando esos datos podríamos saber cuales han reducido o aumentado su puntuación respecto a la media, sabiendo si han ido a peor o a mejor a lo largo del tiempo. 

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(lasVegas[['latitud', 'longitud', 'punt_med_ult_mes', 'count', 'nombre']], lat="latitud", lon="longitud", color_continuous_scale=px.colors.sequential.OrRd, color='punt_med_ult_mes', size='count',  size_max=15, zoom=10, text='nombre', opacity=1)
fig.show()
```

### 2.2. ¿Qué negocios son tendencia en el último mes en todo el mapa?

La filosofía a seguir es la misma, la única diferencia es que ahora no vamos a quedarnos con un solo estado.

Ya tenemos las opiniones del último mes de todo el conjunto en 'opiniones'. Solo tenemos que hacer el mismo proceso que antes, conteo de opiniones y obtención de media respecto a todas. Y por último, unirlo al conjunto de negocios para obtener información de todos los negocios.

```python
opiniones
```

```python
opi_global = opiniones.rename(columns = {'puntuacion_opi':'punt_med_ult_mes'})
punt_med = opi_global.groupby('negocio_id').mean()['punt_med_ult_mes'].to_frame()

count_opi = opi_global.groupby('negocio_id').count()['opinion_id'].to_frame()
count_opi = count_opi.rename(columns = {'opinion_id':'count'})

opi_global = pd.merge(punt_med, count_opi, on='negocio_id', how='inner')
```

```python
opi_global
```

```python
opi_global = pd.merge(opi_global, negocios, on='negocio_id', how='inner')
opi_global
```

```python
opi_global = opi_global.sort_values(['count', 'punt_med_ult_mes'], ascending=False).head(20)
opi_global.head(5)
```

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(opi_global[['latitud', 'longitud', 'punt_med_ult_mes', 'count', 'nombre']], lat="latitud", lon="longitud", color_continuous_scale=px.colors.sequential.OrRd, color='punt_med_ult_mes', size='count',  size_max=15, zoom=10, text='nombre', opacity=1)
fig.show()
```

### Resultados

 Los resultados obtenidos coinciden casi por completo con el análisis que hemos hecho antes sobre Las Vegas, parece ser que los mejores negocios (restaurantes por lo que aparece) se encuentran en el estado de North Las Vegas, la única diferencia con lo anterior es que entre los mejores se cuentran algunos negocios ubicados en Arizona, en concreto, en Phoenix (De los 20 mejores, únicamente dos de ellos). 
 
 
Por lo que concretamente, los mejores negocios (restaurantes en este caso) teniendo en cuenta la tendencia entre las fechas ya indicadas se encuentran en Las Vegas y en Phoenix.

## 3. ¿Cuáles son los peores negocios?

Para hacerlo, tendremos en cuenta los negocios y opiniones tal y como lo hemos hecho antes (continuamos con datos sobre el último mes), solo que en esta ocasion vamos a hacer de forma más manual, trataremos de ver como se distribuyen los datos para saber a partir de cuantas opiniones podemos considerar que son suficientes para valorar algo como malo, y lo mismo para la puntuación.

```python
opiniones
```

```python
opi_global = opiniones.rename(columns = {'puntuacion_opi':'punt_med_ult_mes'})
punt_med = opi_global.groupby('negocio_id').mean()['punt_med_ult_mes'].to_frame()

count_opi = opi_global.groupby('negocio_id').count()['opinion_id'].to_frame()
count_opi = count_opi.rename(columns = {'opinion_id':'count'})

opi_global = pd.merge(punt_med, count_opi, on='negocio_id', how='inner')
```

```python
opi_global
```

Visualizando en primer lugar un describe del total de opiniones para el último mes, podemos ver como la gran mayoría de los negocios no tienen apenas opiniones, por lo que decidimos hacer describe del dataset cortando según los valores que nos han parecido oportunos, en este caso, negocios que tengan mas de 10 opiniones y con una puntuación inferior a 3.

```python
opi_global = pd.merge(opi_global, negocios, on='negocio_id', how='inner')
opi_global.describe()
```

```python
opi_global[(opi_global['count'] > 10) & (opi_global['punt_med_ult_mes'] < 3.0)].describe()
```

```python
opi_global = opi_global[(opi_global['count'] > 10) & (opi_global['punt_med_ult_mes'] < 3.0)]
```

```python
opi_global
```

A partir de los negocios con las condiciones anteriores, colocamos en un mapa los negocios para ver si se sigue algún patron entre los peores negocios.

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(opi_global[['latitud', 'longitud', 'punt_med_ult_mes', 'count', 'nombre']], lat="latitud", lon="longitud", color_continuous_scale=px.colors.sequential.OrRd, color='punt_med_ult_mes', size='count',  size_max=15, zoom=10, text='nombre', opacity=1)
fig.show()
```

### Resultado

A primera vista, aparecen los negocios filtrados en las zonas que más negocios tienen, como parecería lógico (Las Vegas, Phoenix y Toronto principalmente). Solo como curiosidad, en Toronto estos negocios parecen ser teterías y panaderías (o negocios de pasteles). En el caso de Arizona, no parece haber ningún patrón.
