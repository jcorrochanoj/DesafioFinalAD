## 1. Análisis de usuarios 'Influencer'

### 1.1. Obtención de influencers

```python
import pandas as pd
df_opiniones = pd.read_csv("../Dataset/datos_opiniones_sintexto.csv")
influencers = pd.read_csv("../Dataset/influencers.csv")
```

Renombramos columnas ya que se repiten nombres entre datasets, así estamos seguros de lo que significa cada atributo.

```python
df_opiniones = df_opiniones.rename(columns={'util':'util_opi',
                          'guay':'guay_opi', 'puntuacion':'puntuacion_opi'})
df_opiniones
```

```python
influencers = influencers.rename(columns={'util':'util_user',
                          'guay':'guay_user'})
influencers
```

### Unión de dos conjuntos (I)

Hacemos una unión de los conjuntos de usuarios influencer, que para recordar, eran los usuarios pertenecientes a la categoría 3 de los clusters que hicimos y el conjunto de opiniones, quedándonos solo con las opiniones de usuarios influencers (inner join).

```python
df = pd.merge(df_opiniones, influencers, on='usuario_id', how='inner')
```

```python
df
```

### Agrupación
Agrupamos todo el conjunto por usuario y lo ordenamos por su numero de opiniones, su numero de util, divertido y guay. Así consideramos que estos son los mejores 10 influencers.

```python
tenInfluencers = df.groupby('usuario_id').agg({'opinion_id':['count'],
                                'util_opi':['sum'],'divertido':['sum'],'guay_opi':['sum'],
                               'puntuacion_opi':['mean']}).sort_values([('opinion_id', 'count'), ('util_opi', 'sum'), ('divertido', 'sum'), ('guay_opi', 'sum')], ascending=False).head(10)

tenInfluencers
```

Obtenemos un conjunto de usuarios que se puede observar que tienen un gran numero de opiniones y valoraciones (util, divertido...), aunque **cada uno destaca por algo**, sobre todo los dos primeros. **El primer influencer tiene muchas reviews y el segundo lo que tiene son muchos likes.**

Una de las cosas interesantes que podemos hacer es visualizar la localizacion de los negocios a los que han hecho review, ya que la mayoría de opiniones de un usuario rondarán la zona en la que vive. Además de que al ser influencers probablemente acudan a negocios importantes o conocidos. Para ello, nos quedaremos con los id de los negocios a los que han ido nuestros influencers y de nuevo lo uniremos con el dataset de negocios.

```python
df_negocios = pd.read_csv("../Dataset/datos_negocios.csv")
```

```python
df_negocios
```

```python
df_negocios = df_negocios.rename(columns={'puntuacion':'puntuacion_negocio'})
```

### Unión de dos conjuntos (II)

De nuevo, unimos el conjunto que ya tenemos (influencers y sus opiniones) con el dataset de negocios.

```python
df = pd.merge(df, df_negocios, on='negocio_id', how='inner')
```

### 1.4. Redes de amistad de los influencer

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
fiveInfluencersID = tenInfluencers.head(5).index.values.tolist()
fiveInfluencersID
```

```python
fiveInfluencersData = influencers[ influencers['usuario_id'].isin(fiveInfluencersID)]
fiveInfluencersData = fiveInfluencersData[['usuario_id', 'amigos']]
fiveInfluencersData
```

Nos quedamos con los amigos que tiene cada usuario y los NIAS que componen el grafo.

- Edges: Lista de tuplas que contienen los ID de las dos personas que son amigas. Por ejemplo (CxDOIDnH8gp9KXzpBHJYXw, mrru6hbeQLvagU4vkfuqmw)

- Nodos: Lista de ID que representan a cada usuario inmerso en el grafo.

```python
nodes = []
edges = []
for index, row in fiveInfluencersData.iterrows():
    # Obtenemos la lista de amigos
    amigos = row['amigos'].replace(" ", "").split(',')
    edgesAmigos = []
    # Introducimos el ID del influencer en los nodos
    nodes.append(row['usuario_id'])
    # Recorremos los amigos del influencer
    for index, amigo in enumerate(amigos):
        # Introducimos la arista con el nombre del influencer y su amigo
        edgesAmigos.append((row['usuario_id'], amigo))
        # Introducimos el ID del amigo
        nodes.append(amigo)
        if index > 1000:
            break
    edges.extend(edgesAmigos)
        
edges
        
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

Una vez hemos incluido todas las aristas y todos los usuarios en el grafo, los recorremos y les ponemos la características que vemos oportunas.

```python
# Colores que se les va a poner a cada influencer.
coloresInfluencers = {'CxDOIDnH8gp9KXzpBHJYXw': '#FFA15A','bLbSNkLggFnqwNNzzq-Ijw': '#FF0000','PKEzKWv_FktMm2mGPjwd0Q': '#AB63FA', 'ELcQDlf69kb-ihJfxZyL0A': '#00F6FF','DK57YibC5ShBmqQl97CKog': '#00CC96'}

colores = []
size = []
for elem in G.nodes:
    # Si el elemento es un influencer
    if elem in fiveInfluencersID:
        # Se selecciona el color del influencer
        colores.append(coloresInfluencers[elem])
        # Se aumenta el tamaño del influencer.
        size.append(80)
    else:
        colores.append('#DB7093')
        size.append(10)
```

La leyenda de colores sigue siendo la misma que se utilizó en la creación de los mapas de localización. Leyenda:

- **CxDOIDnH8gp9KXzpBHJYXw**: Naranja. Reviews en Toronto.
- **bLbSNkLggFnqwNNzzq-Ijw**: Rojo. Reviews en Las Vegas.
- **PKEzKWv_FktMm2mGPjwd0Q**: Morado. Reviews en las Vegas.
- **ELcQDlf69kb-ihJfxZyL0A**: Azul claro. Mayoría de reviews en Phoenix y unas pocas en Las Vegas.
- **DK57YibC5ShBmqQl97CKog**: Verde. Reviews en Phoenix.

```python
fig = plt.figure(figsize=(11, 11))
nx.draw(G, node_size=size, width=0.2, node_color=colores)
```

Gracias a este grafo, se pueden obtener algunas conclusiones bastante interesantes:

- Como vimos en los mapas en los que mostrábamos la localización de las opiniones de los influencers, **hay varios que se encuentran en Las Vegas**. En concreto, estos usuarios son el **morado**, el **rojo** y el **azul claro**, este último en menor medida. Como se puede observar en el grafo, **existen bastantes personas que son amigos tanto del rojo como del morado**, lo cual concuerda con que visiten la misma ciudad de forma asidua. Esto se puede ver en menor medida en el caso del azul puesto que solo visita ese lugar de vez en cuando (como se puede observar en los mapas)


- En la red se puede observar que el naranja está bastante separado del resto, es decir, no tiene casi ningún amigo en común con el resto. **Este usuario es el que vive en Toronto**, una ciudad bastante alejada de Phoenix y de Las Vegas. Por este motivo, la cantidad de amigos que tienen en común no es tan grande, la gente que es amiga del naranja no está interesada en los negocios que visitan los otros influencers.


- El influencer verde tiene bastantes amigos en común con el azul ya que viven en la misma ciudad.

## 1.5. ¿De donde son los amigos de los Influencers?

Sería ideal tener una lista de los seguidores de cada usuario, así podriamos conocer quienes son estos seguidores y poder relacionar influencer con sus seguidores, los negocios a los que van... etc. En este análisis intentaremos mostrar qué podríamos hacer con esta información, pero lo haremos con los amigos de los Influencers.

```python
import pandas as pd
opiniones = df_opiniones
influencers = pd.read_csv("../Dataset/influencers.csv")
negocios = df_negocios
```

```python
influencers = influencers.rename(columns={'util':'util_user',
                          'guay':'guay_user'})

```

### ¿Qué necesitamos?

Es necesario saber de qué influencer obtener los amigos. Ahora veremos con cuál de los influencers nos quedamos.
Después de obtener la lista de amigos, necesitaremos saber de qué negocios han hecho opiniones para conocer la ubicación de los lugares a los que van. Por último, graficaremos los resultados y sacaremos conclusiones.


### 1: Obtener el id del influencer al que queremos analizar sus amigos

```python
df = pd.merge(opiniones, influencers, on='usuario_id', how='inner')
df
```

```python
tenInfluencers = df.groupby('usuario_id').agg({'opinion_id':['count'],
                                'util_opi':['sum'],'divertido':['sum'],'guay_opi':['sum'],
                               'puntuacion_opi':['mean']}).sort_values([('opinion_id', 'count'), ('util_opi', 'sum'), ('divertido', 'sum'), ('guay_opi', 'sum'), ('puntuacion_opi', 'mean')], ascending=False).head(10)

tenInfluencers
```

```python
tenInfluencers = tenInfluencers.index.tolist()
influencers[influencers['usuario_id'].isin(tenInfluencers)]
```

Nos quedamos con el influencer de otras veces, el más alto en el ranking que hemos hecho, no es el que más amigos tiene pero nos puede servir:

CxDOIDnH8gp9KXzpBHJYXw

### 2: Guardamos sus amigos en una lista

```python
amigosJennifer = influencers[influencers['usuario_id'] == "CxDOIDnH8gp9KXzpBHJYXw"]['amigos'].tolist()[0].replace(' ', '').split(',')
amigosJennifer
```

### 3: Nos quedamos con las opiniones de los amigos

Guardamos las opiniones de los amigos y las agrupamos por negocio, asi no tendremos negocios repetidos y podremos graficar teniendo en cuenta la frecuencia de opiniones de un negocio.

```python
opiniones
```

```python
opinionesAmigos = opiniones[opiniones['usuario_id'].isin(amigosJennifer)]
opinionesAmigos = opinionesAmigos.rename(columns = {'opinion_id':'count'})
opinionesAmigos = opinionesAmigos.groupby('negocio_id').count()['count']
opinionesAmigos = opinionesAmigos.to_frame()
```

### 4: Juntamos los id de los negocios de esas opiniones

El objetivo es obtener la ubicacion de los negocios.

```python
opinionesAmigos = pd.merge(negocios, opinionesAmigos, on='negocio_id', how='inner')
opinionesAmigos
```

### 5: Graficamos los datos en un mapa - CONCLUSIONES

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(opinionesAmigos[['latitud', 'longitud', 'count']], lat="latitud", lon="longitud", color_continuous_scale=px.colors.sequential.Burg, color='count',  size='count',  size_max=15, zoom=5, opacity=0.8)
fig.show()
```

Como se puede ver en el mapa, los amigos de nuestra Influencer Jennifer, de Toronto, están repartidos por EEUU, pero sobre todo se encuentran en Toronto, al igual que ella. Algo a tener en cuenta es que estas ubicaciones son las ubicaciones de los negocios a los que van sus amigos, no los lugares en los que viven (aunque probablemente coincidan en la mayoría de los casos).

Esta información podría ser relevante para proponer a nuestro influencer campañas publicitarias de negocios para la ciudad que nos interesa, en este caso, Toronto. Sabemos que si el el influencer publicita un lugar de Toronto, esta campaña llegará a usuarios que suelen hacer revisiones en Toronto, por lo que estaríamos aprovechando bien esa publicidad.

También hay cabida para las campañas de publicidad de otros productos, por ejemplo, si sabemos la red de amigos y seguidores que tiene este usuario y también sus posibles ubicaciones, podemos lanzar campañas que publicite el usuario y sabemos que van a llegar a x usuarios, por lo que podemos reforzar la venta en esos puntos.

Es por esto por lo que hemos dicho inicialmente que esto sería interesante verlo para los seguidores, pero no disponemos de esa información.
