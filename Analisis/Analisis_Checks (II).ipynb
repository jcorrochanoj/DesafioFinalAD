## 1. Obtención de revisiones ordenadas

A continuación ordenamos el número de revisiones de forma descendiente para poder ver los valores máximos

```python
rev_max=df_check.sort_values(by=['num_revisiones'], ascending=False)
rev_max
```

Como el formato de segundos es de tipo datetime.delta y eso dificultaría operar con los datos o en algunos casos lo imposibilitaría, se ha decidido convertirlo a segundos para poder trabajar con ellos.

```python
df_segundos=df_check
def segundos(x):
    if x != 0:
        seg_total=(x.days*24*60*60)+(x.seconds)
        return seg_total
    elif x == 0:
        return 0
df_segundos['segundos']=df_segundos['media'].apply(lambda x:segundos(x))
```

Como vamos a analizar las medias de los negocios. Los negocios que no tienen media no nos son de utilidad y por lo tanto los eliminamos

```python
df_segundos=df_segundos.drop(df_segundos[df_segundos.segundos==0].index)
```

Una vez hecho esto tratamos de realizar una gráfica del tipo "scatter plot", para ver la distribucion del numero de revisiones con respecto a la media.  Lo que se puede observar es que la mayoría de los datos se acumulan en una franja menor a las 20000 revisiones, por lo tanto hemos dedicido partir el análisis en multiples partes para poder hacerlo más manejable.

```python
rev_graf=df_segundos.drop(['negocio_id','fechas', 'media'],axis=1)
rev_graf.plot.scatter(x= 'num_revisiones', y= 'segundos', c='Darkblue')
```

```python
rev_graf=df_segundos.drop(['negocio_id','fechas', 'media'],axis=1)
rev_graf.plot.scatter(x= 'num_revisiones', y= 'segundos', c='Darkblue')
```

## 2. Negocios con menos de 40 revisiones

Esta reducción en los datos nos permite hacer observaciones más locales pero más detalladas en nuestra opinión.


```python
df_menor40=df_segundos.drop(df_segundos[df_segundos.num_revisiones>40].index)
```

```python
rev_graf=df_menor40.drop(['negocio_id','fechas', 'media', 'num_revisiones'],axis=1)
rev_graf.reset_index().plot.scatter(x='index', y= 'segundos', c='Darkblue')
```

```python

rev_graf=df_menor40.drop(['negocio_id','fechas', 'media'],axis=1)
rev_graf.plot.scatter(x= 'num_revisiones', y= 'segundos', c='Darkblue')
```

En la gráfica mostrada se pueden ver todos los negocios que tienen menos de 40 revisiones y como se distribuyen sus medias. Se puede ver que la mayoria de los datos estan por debajo  de los 3 años (1e8 segundos).

Se puede ver que la gran mayoría del numero de revisiones por negocio se encuentra en la franja de entre 1 y 10 revisiones y que cuantas más revisiones, menor es el tiempo medio que tardan estas en realizarse.
Esto seguramente sea debido los negocios que tienen muchas revisiones, estas se hayan realizado de manera periodica y rutinaria mientras que para pocas revisiones pudieron ser casos mas puntuales y dispersos lo que hace que la media aumente.

A continuación realizamos un boxplot del número de revisiones en el que se puede observar que la distribucion coincide con la anteriormente observada donde se concentran los elementos por debajo de 20 revisiones, siendo 7 la media de todas estas.

```python
rev_box=df_menor40.drop(['negocio_id','fechas', 'media', 'segundos'],axis=1)
rev_box.boxplot(figsize=(10,10))
```

```python
rev_box=df_menor40.drop(['negocio_id','fechas', 'media', 'num_revisiones'],axis=1)
rev_box.boxplot(figsize=(10,10))
rev_box.describe()
```

Mientras que en el boxplot de media de tiempo, se observa que la media entre de tiempo entre revisiones es de 193 dias pero con una desviacion tipica muy grande por lo que los valores no se establecen precisamente en la media.

## 4. Negocios con mas de 5000 revisiones
A continuación se ha decidido realizar un analisis de los negocios con número de revisiones mayor a 5000.

Se puede ver como a diferencia de los dos anteriores intervalos este muestra una distribucion bastante distribuida aunque con poca concentracion de datos.

```python
rev_graf=df_mayor20k.drop(['negocio_id','fechas', 'media'],axis=1)
rev_graf.plot.scatter(x= 'num_revisiones', y= 'segundos', c='Darkblue')
```

Se puede observar que aunque dispongamos de pocos datos la tendencia que sigue es similar a las otras dos mostradas anteriormente aunque el rango de medias para estos negocios con tantas revisiones es de 16 horas

Podemos observar que la media a partir de este valor esta más cercana a 30000 segundos por lo tanto hay más número de revisiones por debajo de los 30000 segundos que por encima.

```python
rev_box=df_mayor20k.drop(['negocio_id','fechas', 'media', 'segundos'],axis=1)
rev_box.boxplot(figsize=(10,10))
```

Como se tienen tan pocos datos, con el número de revisiones este este boxplot no se tienen demasiados puntos atípicos. Aunque sería interesante concentrarnos en estos puntos atípicos, investigarlos y ver cuáles son sus propiedades.

```python
rev_box=df_mayor20k.drop(['negocio_id','fechas', 'media', 'num_revisiones'],axis=1)
rev_box.boxplot(figsize=(10,10))
```

A diferencia del boxplot de la media de tiempo anterior este no muestra puntos atipicos lo que nos hace pensar que  las medias de los negocios están dispuestas de una manera más uniforme.

```python
a=df_mayor20k.sort_values(by=['num_revisiones'], ascending=False).head(3)
df=pd.merge(a,df_negocios, on='negocio_id', how='inner')
df
```

Estos son los 3 negocios que mas revisiones tienen y se puede ver que reciben una revision por hora o cada media hora.

Uniendo el dataframe de negocios y el de checks para sacar el nombre de estos negocios obtenemos que los negocios son aeropuertos en los que se tienen que realizar revisiones muy frecuentemente, en concreto, cada media hora para el de McCarran en Las Vegas.

Teniendo en cuenta que los servicios que presta estos negocios son los de parking, transporte, seguridad... etc, es lógico pensar que estos tengan muchas revisiones físicas ya que todo debe ser correcto.
