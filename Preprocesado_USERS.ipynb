### Búsqueda de valores nulos

```python
print("¿Hay valores nulos?\n", pd.isnull(df_users).any(), "\n")
print("¿Cuantos?\n", pd.isnull(df_users).sum(), "\n")
```

Se han encontrado algunos valores nulos en el atributo nombre y muchísimos en el atributo "experto". En este último atributo se tienen los años en el que el usuario ha sido experto, por lo que muchos usuarios no han sido nunca expertos.

Decidimos completar los nombres nulos de nombre con "SinNombre" y los valores nulos de experto con "Ninguno". No eliminamos las instancias de usuarios sin nombre porque podemos seguir identificándolo con su ID. Tambien cambiar None de un usuario sin amigos por "SinAmigos", para tener controlado e identificado ese caso.

```python
# Rellenamos valores nulos y NaN con un valor específico y controlado
df_users['experto'] = df_users['experto'].fillna("Ninguno")
df_users['nombre'] = df_users['nombre'].fillna("SinNombre")
df_users['amigos'] = df_users['amigos'].replace("None", "SinAmigos")
df_users
```

```python
print("¿Hay valores nulos?\n", pd.isnull(df_users).any(), "\n")
print("¿Cuantos?\n", pd.isnull(df_users).sum(), "\n")
```

### Búsqueda de inconsistencias en los datos

Algo que nos ha parecido interesante comprobar es si los amigos de cada usuario están en la base datos, hemos decidido hacer esta comprobación porque nos parecía algo que debía ser consistente.

```python
users_id = df_users["usuario_id"]

# Recorremos las listas de amigos de cada usuario para comprobar si existen
for amigos in df_users["amigos"][0:1]:
    listaAmigos = amigos.replace(" ", "")
    listaAmigos = listaAmigos.split(",")
    numAmigos = len(listaAmigos)
    
    exists = df_users["usuario_id"].isin(listaAmigos)
    
    print("Numero de amigos: " + str(numAmigos))
    print(exists.value_counts())
```

Comprobamos solo para el primer usuario ya que imprimir todos es demasiado, pero ocurre lo mismo con todos. En el caso impreso, el usuario tiene 99 amigos de los cuales solo 13 están en la base de datos. Ya que inicialmente esto nos parece muy extraño, decidimos imprimir cuales están y cuales no para comprobarlo manualmente al menos para un par de ejemplos:

```python
users_id = df_users["usuario_id"]

# Recorremos las listas de amigo de cada usuario imprimiendo para cada user si existe o no
for amigos in df_users["amigos"][0:1]:
    listaAmigos = amigos.replace(" ", "")
    listaAmigos = listaAmigos.split(",")
    numAmigos = len(listaAmigos)
    for amigo in listaAmigos:
        lista = []
        lista.append(amigo)
        exists = df_users["usuario_id"].isin(lista).to_numpy()

        if True in exists:
            print(amigo + ": True")
        else:
            print(amigo + ": False")
```

### Comprobación sintáctica de la fecha

Comprobamos si la fecha es correcta sintácticamente utilizando una expresión regular. No se imprime ninguna fecha incorrecta, por lo que todas están bien.

```python
import re
pattern = re.compile("[0-9][0-9][0-9][0-9][-][0-1][0-9][-][0-3][0-9]\s[0-2][0-9][:][0-5][0-9][:][0-5][0-9]")


def preprocesa(x):
    if(pattern.match(x) == None):
        print("date not matching: " + str(x))
        return True
    return False

df_users['usuario_desde'].apply(lambda x: preprocesa(x))
```

### Transformación de los datos
De todos los elementos que podemos quedarnos de la fecha, decidimos quedarnos con el día, el mes, el añoy la hora ya que puede ser que posteriormente nos sean útiles. El atributo que puede llegar a ser más controversial es el minuto puesto que se puede considerar que es hilar demasiado fino. Los atributos con los que nos quedamos, nos pueden servir para hacer análisis de los registros que se realizan. Hemos tenido dudas a la hroa de quedarnos con la hora, pero nos podría servir, por ejemplo, para determinar a qué hora se realizan más registros e informar a nuestro cliente (podría ser necesario un refuerzo en los servidores a una hora determinada).

```python
# Dividimos la usuario_desde en año, mes , dia, hora y minuto
df_users['anyo_reg'] = pd.DatetimeIndex(df_users['usuario_desde']).year
df_users['mes_reg'] = pd.DatetimeIndex(df_users['usuario_desde']).month
df_users['dia_reg'] = pd.DatetimeIndex(df_users['usuario_desde']).day
df_users['hora_reg'] = pd.DatetimeIndex(df_users['usuario_desde']).hour

# Eliminamos el atributo usuario_desde
df_users = df_users.drop("usuario_desde", axis=1)
df_users
```

### Número de amigos

Aunque tenemos en un array qué amigos tiene un usuario y si no tiene amigos, vamos a incluir una nueva columna cuyo objetivo sea indicarnos la cantidad concreta de amigos que tiene cada usuario.

```python
def genNumAmigos (row):
    amigos = row['amigos'].split(',')
    numAmigos = len(amigos)
    if numAmigos == 1 and amigos[0] == 'SinAmigos':
        numAmigos = 0
        
    return numAmigos

df_users["num_amigos"] = df_users.apply(lambda row: genNumAmigos(row), axis = 1)
```

```python
cols = df_users.columns.tolist()
cols = cols[:8] + cols[-1:] + cols[8:-1]

df_users = df_users[cols]
```
