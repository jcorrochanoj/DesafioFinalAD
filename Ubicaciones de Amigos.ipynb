## ¿De donde son los amigos de los Influencers?

Sería ideal tener una lista de los seguidores de cada usuario, así podriamos conocer quienes son estos seguidores y poder relacionar seguidores, los negocios a los que van... etc. En este análisis intentaremos mostrar qué podríamos hacer con esta información, pero lo haremos con los amigos de los Influencers.

```python
import pandas as pd
opiniones = pd.read_csv("Dataset/datos_opiniones_sintexto.csv")
influencers = pd.read_csv("Dataset/influencers.csv")
negocios = pd.read_csv("Dataset/datos_negocios_limpios.csv")
```

```python
opiniones = opiniones.rename(columns={'util':'util_opi',
                          'guay':'guay_opi', 'puntuacion':'puntuacion_opi'})

influencers = influencers.rename(columns={'util':'util_user',
                          'guay':'guay_user'})

```

### ¿Qué necesitamos?

Es necesario saber de qué influencer obtener los amigos. Ahora veremos con cuál de los influencers nos quedamos.
Después de obtener la lista de amigos, necesitaremos saber de qué negocios han hecho opiniones para conocer la ubicación de los lugares a los que van. Por último, graficaremos los resultados y sacaremos conclusiones.


### 1: Obtener el id del influencer al que queremos analizar sus amigos

```python
df = pd.merge(opiniones, influencers, on='usuario_id', how='inner')
df
```

```python
tenInfluencers = df.groupby('usuario_id').agg({'opinion_id':['count'],
                                'util_opi':['sum'],'divertido':['sum'],'guay_opi':['sum'],
                               'puntuacion_opi':['mean']}).sort_values([('opinion_id', 'count'), ('util_opi', 'sum'), ('divertido', 'sum'), ('guay_opi', 'sum'), ('puntuacion_opi', 'mean')], ascending=False).head(10)

tenInfluencers
```

```python
tenInfluencers = tenInfluencers.index.tolist()
influencers[influencers['usuario_id'].isin(tenInfluencers)]
```

Nos quedamos con el influencer de otras veces, el más alto en el ranking que hemos hecho, no es el que más amigos tiene pero nos puede servir:

CxDOIDnH8gp9KXzpBHJYXw

### 2: Guardamos sus amigos en una lista

```python
amigosJennifer = influencers[influencers['usuario_id'] == "CxDOIDnH8gp9KXzpBHJYXw"]['amigos'].tolist()[0].replace(' ', '').split(',')
amigosJennifer
```

### 3: Nos quedamos con las opiniones de los amigos

Guardamos las opiniones de los amigos y las agrupamos por negocio, asi no tendremos negocios repetidos y podremos graficar teniendo en cuenta la frecuencia de opiniones de un negocio.

```python
opiniones
```

```python
opinionesAmigos = opiniones[opiniones['usuario_id'].isin(amigosJennifer)]
opinionesAmigos = opinionesAmigos.rename(columns = {'opinion_id':'count'})
opinionesAmigos = opinionesAmigos.groupby('negocio_id').count()['count']
opinionesAmigos = opinionesAmigos.to_frame()
```

### 4: Juntamos los id de los negocios de esas opiniones

El objetivo es obtener la ubicacion de los negocios.

```python
opinionesAmigos = pd.merge(negocios, opinionesAmigos, on='negocio_id', how='inner')
opinionesAmigos
```

### 5: Graficamos los datos en un mapa - CONCLUSIONES

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(opinionesAmigos[['latitud', 'longitud', 'count']], lat="latitud", lon="longitud", color_continuous_scale=px.colors.sequential.Burg, color='count',  size='count',  size_max=15, zoom=5, opacity=0.8)
fig.show()
```

Como se puede ver en el mapa, los amigos de nuestra Influencer Jennifer, de Toronto, están repartidos por EEUU, pero sobre todo se encuentran en Toronto, al igual que ella. Algo a tener en cuenta es que estas ubicaciones son las ubicaciones de los negocios a los que van sus amigos, no los lugares en los que viven (aunque probablemente coincidan en la mayoría de los casos).

Esta información podría ser relevante para proponer a nuestro influencer campañas publicitarias de negocios para la ciudad que nos interesa, en este caso, Toronto. Sabemos que si el el influencer publicita un lugar de Toronto, esta campaña llegará a usuarios que suelen hacer revisiones en Toronto, por lo que estaríamos aprovechando bien esa publicidad.

También hay cabida para las campañas de publicidad de otros productos, por ejemplo, si sabemos la red de amigos y seguidores que tiene este usuario y también sus posibles ubicaciones, podemos lanzar campañas que publicite el usuario y sabemos que van a llegar a x usuarios, por lo que podemos reforzar la venta en esos puntos.

Es por esto por lo que hemos dicho inicialmente que esto sería interesante verlo para los seguidores, pero no disponemos de esa información.
