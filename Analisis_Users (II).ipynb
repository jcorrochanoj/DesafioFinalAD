## Análisis de usuarios 'Influencer'

```python
import pandas as pd
df_opiniones = pd.read_csv("Dataset/datos_opiniones_sintexto.csv")
influencers = pd.read_csv("Dataset/influencers.csv")
```

Renombramos columnas ya que se repiten nombres entre datasets, así estamos seguros de lo que significa cada atributo.

```python
df_opiniones = df_opiniones.rename(columns={'util':'util_opi',
                          'guay':'guay_opi', 'puntuacion':'puntuacion_opi'})
df_opiniones
```

```python
influencers = influencers.rename(columns={'util':'util_user',
                          'guay':'guay_user'})
influencers
```

### Unión de dos conjuntos (I)

Hacemos una unión de los conjuntos de usuarios influencer, que para recordar, eran los usuarios pertenecientes a la categoría 3 de los clusters que hicimos y el conjunto de opiniones, quedándonos solo con las opiniones de usuarios influencers (inner join).

```python
df = pd.merge(df_opiniones, influencers, on='usuario_id', how='inner')
```

```python
df
```

### Agrupación
Agrupamos todo el conjunto por usuario y lo ordenamos por su numero de opiniones, su numero de util, divertido y guay y por su puntuación. Así consideramos que estos son los mejores 10 influencers.

```python
tenInfluencers = df.groupby('usuario_id').agg({'opinion_id':['count'],
                                'util_opi':['sum'],'divertido':['sum'],'guay_opi':['sum'],
                               'puntuacion_opi':['mean']}).sort_values([('opinion_id', 'count'), ('util_opi', 'sum'), ('divertido', 'sum'), ('guay_opi', 'sum'), ('puntuacion_opi', 'mean')], ascending=False).head(10)

tenInfluencers
```

Obtenemos un conjunto de usuarios que se puede observar que tienen un gran numero de opiniones y valoraciones (util, divertido...), aunque **cada uno destaca por algo**, sobre todo los dos primeros. **El primer influencer tiene muchas reviews y el segundo lo que tiene son muchos likes.**

Una de las cosas interesantes que podemos hacer es visualizar la localizacion de los negocios a los que han hecho review, ya que la mayoría de opiniones de un usuario rondarán la zona en la que vive. Además de que al ser influencers probablemente acudan a negocios importantes o conocidos. Para ello, nos quedaremos con los id de los negocios a los que han ido nuestros influencers y de nuevo lo uniremos con el dataset de negocios.

```python
df_negocios = pd.read_csv("Dataset/datos_negocios.csv")
```

```python
df_negocios
```

```python
df_negocios = df_negocios.rename(columns={'puntuacion':'puntuacion_negocio'})
```

### Unión de dos conjuntos (II)

De nuevo, unimos el conjunto que ya tenemos (influencers y sus opiniones) con el dataset de negocios.

```python
df = pd.merge(df, df_negocios, on='negocio_id', how='inner')
```

```python
df
```

### ¿Dónde hacen sus reviews los influencers?

Como primer acercamiento, vamos a estudiar únicamente los negocios del primer influencer que hemos encontrado. El id de este usuario es "CxDOIDnH8gp9KXzpBHJYXw". Para ello, nos quedaremos solo con las instancias del dataset completo anterior (negocios, opiniones y usuarios) para este usuario y graficaremos las ubicaciones en un mapa.

```python
best_influencer = df[df['usuario_id'] == "CxDOIDnH8gp9KXzpBHJYXw"]
```

```python
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(best_influencer[['latitud', 'longitud']], lat="latitud", lon="longitud",
                  color_discrete_sequence=["#FFA15A"], size_max=15, zoom=10, opacity=0.5)
fig.show()
```

Tal y como se puede ver en el mapa interactivo o en la foto de abajo, el primer influencer parece ser de Toronto, ya que la gran matoría de sus reviews son de esta zona. Es muy intersante ver que simplemente usando los datos podemos ver graficadas sobre un mapa los distintos negocios formando calles.
