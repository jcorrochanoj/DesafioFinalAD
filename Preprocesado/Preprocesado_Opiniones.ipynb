### BÃºsqueda de valores nulos

```python
print("Â¿Hay valores nulos?\n", pd.isnull(df_opiniones).any(), "\n")
print("Â¿Cuantos?\n", pd.isnull(df_opiniones).sum(), "\n")
```

```python
import numpy as np

# Buscamos las filas en las que hay nulos para comprobar la anomalÃ­a y encontrar explicaciones
filas, _ = np.where(df_opiniones.isnull().values == True)

# Nos quedamos en la lista con las filas que no estan repetidas.
filas = list(dict.fromkeys(filas))

for fila in filas:
    print(str(df_opiniones.iloc[fila]) + "\n")
```

DespuÃ©s de imprimir las filas incorrectas, se pueden hacer varias observaciones:

- **Dos opiniones sin fecha.** Realmente, no se nos ocurre cuÃ¡l puede ser el motivo por el que haya opiniones sin fecha. De todos modos, no es algo determinante en el anÃ¡lisis de datos, asÃ­ que lo que vamos a hacer es **eliminar directamente las filas**. PodrÃ­amos cambiar su valor NaN a un valor controlado pero tendrÃ­amos problemas a la hora de cambiar el formato de las fechas.


- **Dos opiniones con casi todo NaN y valores incorrectos.** Se puede decir que estas son las opiniones mÃ¡s extraÃ±as de todas: el texto de la opinion aparece en opinion_ID y la fecha en usuario_ID. Esto puede deberse a que hay usuarios que no han rellenado correctamente algunos formularios y **no existe el suficiente control en el sistema a la hora de hacer inserciones en la base de datos**. Debido a que son completamente incorrectas, deberÃ­amos eliminar estas instancias por completo.


- **Dos opiniones con texto en NaN.** Aunque carecemos de informaciÃ³n suficiente, podrÃ­a ser que el motivo por el que no hay texto en algunas opiniones fuera porque es un campo opcional. En ese caso, lo Ãºnico que vamos a hacer es sustituir el valor NaN por: **"SinTexto"**

```python
# Rellenamos valores nulos y NaN con un valor especÃ­fico y controlado
df_opiniones['texto'] = df_opiniones['texto'].fillna("SinTexto")
```

```python
import numpy as np

# Buscamos las filas en las que hay nulos para comprobar la anomalÃ­a y encontrar explicaciones
filas, _ = np.where(df_opiniones.isnull().values == True)

# Nos quedamos en la lista con las filas que no estan repetidas.
filas = list(dict.fromkeys(filas))

for fila in filas:
    df_opiniones = df_opiniones.drop(fila, axis=0)
```

Al hacer la limpieza anterior, nos hemos dado cuenta de que hay ID que son incorrectas. Sin embargo, nos hemos percatado de esto debido a que habÃ­a valores NaN en esas mismas instancias. Para comprobar que todos los ID tienen el formato correcto, vamos a hacer unas comprobaciones:

#### ComprobaciÃ³n sintÃ¡ctica de opinion_id

```python
import re
# Expresion regular que indica el formato de los ID
pattern = re.compile("[A-Za-z0-9-_]{22}")

def preprocesa(x):
    if(pattern.match(x) == None):
        print("opinion_id not matching: " + str(x) + "\n")
        return True
    return False

print("Numero de opinion_id incorrectas:" + str(df_opiniones['opinion_id'].apply(lambda x: preprocesa(x)).sum()))
```

#### ComprobaciÃ³n sintÃ¡ctica de negocio_id

```python
import re
# Expresion regular que indica el formato de los ID
pattern = re.compile("[A-Za-z0-9-_]{22}")

def preprocesa(x):
    if not isinstance(x, str):
        print(x)
    if(pattern.match(x) == None):
        print("negocio_id not matching: " + str(x) + "\n")
        return True
    return False

# Aplicamos la funcion preprocesa a todas las filas del dataframe y sumamos los valores positivos.
print("Numero de negocio_id incorrectas:" + str(df_opiniones['negocio_id'].apply(lambda x: preprocesa(x)).sum()))
```

#### ComprobaciÃ³n sintÃ¡ctica de usuario_id

```python
import re
# Expresion regular que indica el formato de los ID
pattern = re.compile("[A-Za-z0-9-_]{22}")

def preprocesa(x):
    if not isinstance(x, str):
        print(x)
    if(pattern.match(x) == None):
        print("usuario_id not matching: " + str(x) + "\n")
        return True
    return False

# Aplicamos la funcion preprocesa a todas las filas del dataframe y sumamos los valores positivos.
print("Numero de usuario_id incorrectas:" + str(df_opiniones['usuario_id'].apply(lambda x: preprocesa(x)).sum()))
```

Como se puede comprobar, no hay ningÃºn ID incorrecto despuÃ©s de haber hecho la limpieza de los valores nulos o NaN.

### ComprobaciÃ³n sintÃ¡ctica de la fecha

```python
import re
pattern = re.compile("[0-9][0-9][0-9][0-9][-][0-1][0-9][-][0-3][0-9]\s[0-2][0-9][:][0-5][0-9][:][0-5][0-9]")

def preprocesa(x):
    if(pattern.match(x) == None):
        print("date not matching: " + str(x))
        return True
    return False

print("Numero de fechas incorrectas: " + str(df_opiniones['fecha'].apply(lambda x: preprocesa(x)).sum()))
```

Todas las fechas son correctas exceptuando las dos que hemos cambiado nosotros a mano por el valor "SinFecha".

### TransformaciÃ³n inicial de la fecha

```python
# Dividimos la fecha en aÃ±o, mes , dia, hora y minuto
df_opiniones['anyo'] = pd.DatetimeIndex(df_opiniones['fecha']).year
df_opiniones['mes'] = pd.DatetimeIndex(df_opiniones['fecha']).month
df_opiniones['dia'] = pd.DatetimeIndex(df_opiniones['fecha']).day
df_opiniones['hora'] = pd.DatetimeIndex(df_opiniones['fecha']).hour

# Eliminamos el atributo fecha
df_opiniones = df_opiniones.drop("fecha", axis=1)
df_opiniones
```

### ComprobaciÃ³n de distribuciÃ³n de valores

Vamos a hacer un describe con el objetivo de encontrar valores que puedan ser incorrectos. En la secciÃ³n de anÃ¡lisis haremos un anÃ¡lisis mÃ¡s exhaustivo.

```python
# Lo ponemos de esta forma para quitar la notaciÃ³n cientÃ­fica
df_opiniones.describe().apply(lambda s: s.apply(lambda x: format(x, 'g')))
```

Como se puede observar en el describe, tanto la columna "util" como la columna "guay" tienen como valor mÃ­nimo el -1. Esto podrÃ­a no ser un error en caso de que se pudiera votar negativamente algunas opiniones. Para comprobar si eso es asÃ­, vamos a obtener cuÃ¡ntas filas tienen esas columnas a -1.

```python
import numpy as np

# Buscamos las filas en las que hay nulos para comprobar la anomalÃ­a y encontrar explicaciones
filas, _ = np.where(df_opiniones[df_opiniones["util"] == -1].values == True)

utilIncorrecto = df_opiniones[df_opiniones["util"] == -1].opinion_id.values[0]
print("ID de la fila con util incorrecto: " + str(utilIncorrecto))
guayIncorrecto = df_opiniones[df_opiniones["guay"] == -1].opinion_id.values[0]
print("ID de la fila con guay incorrecto: " + str(guayIncorrecto))
```

Como se puede comprobar, solo hay una fila con esos valores negativos por lo que podemos deducir que fue un error del sistema. Como no es algo generalizado, vamos a optar por eliminar esa fila.
