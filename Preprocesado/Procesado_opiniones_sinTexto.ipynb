## Dataset de opiniones limpios para obtener sentimiento

### Explicación

En esta parte del preprocesado hemos decidido utilizar la librería textblob para realizar un análisis de sentimiento en la opiniones de los usuarios. Con ello lo que buscamos es reducir el tamaño del fichero y tener una medida de sentimiento que nos permita ir por otros caminos en el análisis de datos desarrollado. Teniendo en cuenta que la librería textblob nos devuelve el sentimiento y la objetividad, pero hemos decidido pasar este último valor a subjetividad puesto que hemos considerado que es más entendible.

```python
import pandas as pd
import textblob as tb
```

```python
def sentiment(x):
    text=tb.TextBlob(x)
    sentimiento=(text.sentiment[0]+1)/2
    subjetividad=text.sentiment[1]
    return(sentimiento, subjetividad)
```

```python
df = pd.read_csv('../Dataset/datos_opiniones_limpio.csv')
df
```

```python
df["sentimiento/subjetivo"]=df["texto"].apply(lambda x: sentiment(x))
```

```python
df=df.drop("texto",axis=1)
```

```python
def objetividad(x):
    return 1-(eval(x)[1])
```

```python
df["sentimiento"]=df["sentimiento/subjetivo"].apply(lambda x: eval(x)[0])
df["objetividad"]=df["sentimiento/subjetivo"].apply(lambda x: objetividad(x))
df=df.drop("sentimiento/subjetivo",axis=1)
```

```python
df.to_csv("../Dataset/datos_opiniones_limpio_sentimientos_sintexto.csv", index=False)
```
